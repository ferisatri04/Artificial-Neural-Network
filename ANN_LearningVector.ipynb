```python
#Introduction to LVQ on-line training algorithm and its implementation using python numpy
#clustering sample data using python library pandas, matplotlib, numpy, 
#and LVQ algorithm implementation
import pandas as pd
import numpy as np
pd.set_option('display.max_rows', 150)
```

```python
#Getting data must be csv file
#reading data as data frame
input_data = pd.read_csv("E:\zoo.csv")
input_data.iloc[:, :]
```

```python
#we take all columns of characteristic and save name_animal and class_type
#there are 16 characteristic (16 columns) of animal and one column number total is 17
#for n data based file
trunc_data = input_data.iloc[:, 1:17]
trunc_data.iloc[:, :]
```

```python
trunc_data.iloc[30:71, :]
```

```python

sorted_data = trunc_data.sort_values('class_type')
sorted_data.iloc[:, :]
```

```python
sorted_data.iloc[30:71, :]
```

```python
ref_vec = [sorted_data.values[i, :] for i in range(0, 4)]
for i in range(41,44):
    ref_vec.append(sorted_data.values[i, :])

ref_vec.append(sorted_data.values[61, :])
for i in range(66, 68):
    ref_vec.append(sorted_data.values[i, :])    

ref_vec.append(sorted_data.values[79, :])

for i in range(83, 85):
    ref_vec.append(sorted_data.values[i, :])

for i in range(91, 93):
    ref_vec.append(sorted_data.values[i, :])
ref_vec = np.array(ref_vec)

print(ref_vec)
```

```python
from LVQ.LVQNet import LVQ
indices = np.arange(trunc_data.values.shape[0])
anim_lvq = LVQ(7,1,16)
anim_lvq.train(trunc_data.values, ref_vec, num_epochs=200, init_learning_rate=0.1)
```

```python
def predict(df):
    bmu, bmu_idx = anim_lvq.find_bmu(df.values, ref_vec)
    df['Class'] = bmu[-1]
    return df
classified_df = trunc_data.apply(predict, axis=1)
classified_df.iloc[:]
```

```python
joined_df = input_data.join(classified_df['Class'])
joined_df = joined_df.sort_values('animal_name')
#joined_df.iloc[:, :]
joined_df[['animal_name', 'class_type', 'Class']]
```

```python
from matplotlib import pyplot as plt
from matplotlib import patches as patches
import matplotlib.lines as mlines

fig = plt.figure(figsize=(10,10))
# setup axes
ax = fig.add_subplot(111)
scale = 1000
ax.set_xlim((0, 7 * scale))
ax.set_ylim((0, scale))
ax.tick_params(labelcolor='w')
ax.set_title("Animal Classification by using LVQ", color = 'y')

for x in range(0, 7):
    for y in range(0, 1):
        ax.add_patch(patches.Rectangle((x*scale, 0), scale, scale,
                     facecolor='white',
                     edgecolor='grey'))
legend_map = {}
        
for index, row in joined_df.iterrows():
    x_cor = np.random.randint((row['Class']-1) * scale, row['Class']*scale)
    y_cor = np.random.randint(0, scale)
    marker = "$\\ " + row['animal_name'][0:4]+"$"
    marker = marker.lower()
    ax.plot(x_cor, y_cor, color=None, marker=marker, markersize=30)
    label = row['animal_name']
    if not label in legend_map:
        legend_map[label] =  mlines.Line2D([], [], color='black', marker=marker, linestyle='None',
                          markersize=25, label=label)
plt.legend(handles=list(legend_map.values()), bbox_to_anchor=(1, 1))
plt.show()
```
